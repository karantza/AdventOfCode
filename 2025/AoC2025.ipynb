## Day 1
tried to be clever in part 2 but only a brute force solution actually got the right answer. the lesson here is to not prematurely optimize

```python
# example input
puz_input = [
"L68",
"L30",
"R48",
"L5",
"R60",
"L55",
"L1",
"L99",
"R14",
"L82"]


file_path = 'in1.txt'
with open(file_path, 'r') as file:
    puz_input = file.readlines()


dial = 50
landed_zeroes = 0
passed_zeroes = 0

def turn(x):
    global landed_zeroes, passed_zeroes
    
    dir_str = x[0:1]
    val = int(x[1:])
    
    d = dial
    
    if dir_str == "L":
        while val > 0:
            # negative
            val -= 1
            d -= 1
            if (d < 0):
                d += 100
            if (d == 0):
                passed_zeroes += 1
    else:
        while val > 0:
            val -= 1
            d += 1
            if (d > 99):
                d -= 100
            if (d == 0):
                passed_zeroes += 1
    
    if d == 0:
        landed_zeroes += 1
    
    return d

for x in puz_input:
    dial = turn(x)
    
print(landed_zeroes, passed_zeroes);

```
